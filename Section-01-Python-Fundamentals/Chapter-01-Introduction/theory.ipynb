Bhai agar tum Python ke `print()` function ka **poora behavior** samajhna chahte ho, to ye main points hain:

## 1. Basic Print

```python
print("Hello World")
```

Output:

```text
Hello World
```

---

## 2. Multiple Values Print

```python
name = "Prahlad"
age = 21

print(name, age)
```

Output:

```text
Prahlad 21
```

Default me values ke beech **space** lagta hai.

---

## 3. `sep` Parameter

Default separator `" "` hota hai.

```python
print("A", "B", "C", sep="-")
```

Output:

```text
A-B-C
```

```python
print("2026", "05", "28", sep="/")
```

Output:

```text
2026/05/28
```

---

## 4. `end` Parameter

Default end `"\n"` (new line) hota hai.

```python
print("Hello", end=" ")
print("World")
```

Output:

```text
Hello World
```

---

## 5. New Line Character

```python
print("Hello\nWorld")
```

Output:

```text
Hello
World
```

---

## 6. Tab Character

```python
print("Name\tAge")
```

Output:

```text
Name    Age
```

---

## 7. Quotes Print Karna

```python
print("I am \"Prahlad\"")
```

Output:

```text
I am "Prahlad"
```

---

## 8. Variables Print

```python
name = "Prahlad"

print(name)
```

Output:

```text
Prahlad
```

---

## 9. f-string (Most Used)

```python
name = "Prahlad"
age = 21

print(f"My name is {name} and age is {age}")
```

Output:

```text
My name is Prahlad and age is 21
```

---

## 10. String Formatting

```python
print("My name is {} and age is {}".format("Prahlad", 21))
```

Output:

```text
My name is Prahlad and age is 21
```

---

## 11. Arithmetic Print

```python
print(10 + 20)
```

Output:

```text
30
```

```python
print(5 * 6)
```

Output:

```text
30
```

---

## 12. List Print

```python
nums = [1, 2, 3]

print(nums)
```

Output:

```text
[1, 2, 3]
```

---

## 13. Dictionary Print

```python
user = {
    "name": "Prahlad",
    "age": 21
}

print(user)
```

Output:

```text
{'name': 'Prahlad', 'age': 21}
```

---

## 14. Object Print Behavior

```python
class Student:
    pass

s = Student()

print(s)
```

Output kuch aisa:

```text
<__main__.Student object at 0x...>
```

Kyuki Python object ka default representation print karta hai.

### `__str__()` Use Karke

```python
class Student:
    def __str__(self):
        return "Prahlad Student"

s = Student()

print(s)
```

Output:

```text
Prahlad Student
```

---

## 15. File Me Print Karna

```python
f = open("data.txt", "w")

print("Hello", file=f)

f.close()
```

Console ki jagah file me likhega.

---

## 16. Flush Output

```python
print("Loading...", flush=True)
```

Buffer ko immediately flush kar deta hai.

---

## 17. Return Value

```python
x = print("Hello")

print(x)
```

Output:

```text
Hello
None
```

Kyuki `print()` **hamesha `None` return karta hai**.

---

## Actual Internal Signature

Python me roughly:

```python
print(
    *objects,
    sep=' ',
    end='\n',
    file=sys.stdout,
    flush=False
)
```

Yahi 5 cheezein `print()` ke core behavior ko control karti hain:

* `*objects` → kya print karna hai
* `sep` → values ke beech kya aayega
* `end` → line ke end me kya aayega
* `file` → kahan print hoga
* `flush` → buffer turant clear karna hai ya nahi

Ye Python ke `print()` ka lagbhag poora behavior hai jo interviews aur real projects me kaam aata hai.
