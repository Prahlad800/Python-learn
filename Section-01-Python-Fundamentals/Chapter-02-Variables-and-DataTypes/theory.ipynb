# 1. Python Output (`print()`)

`print()` screen par output dikhane ke liye use hota hai.

```python
print('Hello World')
```

Output:

```text
Hello World
```

### String Print

```python
print('salman khan')
```

Output:

```text
salman khan
```

### Error Example

```python
print(salman khan)
```

Ye error dega:

```text
SyntaxError
```

Kyuki `salman khan` variable nahi hai aur quotes bhi nahi lage.

### Number Print

```python
print(7)
print(7.7)
```

Output:

```text
7
7.7
```

### Multiple Values

```python
print('Hello',1,4.5,True)
```

Output:

```text
Hello 1 4.5 True
```

Default separator space hota hai.

### Custom Separator

```python
print('Hello',1,4.5,True,sep='/')
```

Output:

```text
Hello/1/4.5/True
```

### New Line

```python
print('hello')
print('world')
```

Output:

```text
hello
world
```

### End Parameter

```python
print('hello',end='-')
print('world')
```

Output:

```text
hello-world
```
