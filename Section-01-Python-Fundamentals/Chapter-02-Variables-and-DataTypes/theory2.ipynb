# Q1. Separator and End

# Q6. Dogs and Chickens

### Theory

Dog = 4 legs

Chicken = 2 legs

Let:

```text
d = dogs
c = chickens
```

Equations:

```text
d + c = heads
4d + 2c = legs
```

### Code

```python
heads = int(input("Enter heads: "))
legs = int(input("Enter legs: "))

dogs = (legs - 2*heads)//2
chickens = heads - dogs

print("Dogs =",dogs)
print("Chickens =",chickens)
```

### Example

Input

```text
heads = 4
legs = 12
```

Output

```text
Dogs = 2
Chickens = 2
```

# Q7. Sum of Squares

### Theory

Formula:

1^2+2^2+3^2+\cdots+n^2

### Code

```python
n = int(input("Enter n: "))

sum_sq = 0

for i in range(1,n+1):
    sum_sq += i*i

print("Sum =",sum_sq)
```

# Q8. Nth Term of Arithmetic Progression

### Theory

Common Difference:

```text
d = second_term - first_term
```

Nth Term:

a_n=a+(n-1)d

### Code

```python
a = int(input("Enter first term: "))
b = int(input("Enter second term: "))
n = int(input("Enter n: "))

d = b - a

nth_term = a + (n-1)*d

print("Nth Term =",nth_term)
```

# Q10. Milk Tank and Glasses

### Theory

Tank Volume:

V_{tank}=L\times B\times H

Glass (Cylinder) Volume:

genui{"math_block_widget_always_prefetch_v2":{"content":"V_{glass}=\pi r^2 h"}}

Number of glasses:

```text
Tank Volume ÷ Glass Volume
```

### Code

```python
import math

H = float(input("Enter tank height: "))
L = float(input("Enter tank length: "))
B = float(input("Enter tank breadth: "))

h = float(input("Enter glass height: "))
r = float(input("Enter glass radius: "))

tank_volume = H * L * B

glass_volume = math.pi * r * r * h

glasses = tank_volume // glass_volume

print("Number of glasses =",int(glasses))
```

### Example

```text
Tank
20 x 20 x 20 = 8000 cm³

Glass
π × 1² × 3
= 9.42 cm³
```

Output:

```text
Number of glasses = 849
```
